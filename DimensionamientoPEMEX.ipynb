```python
class IC_tubo_carcasa:
    def __init__(self,cec,csc,cef,csf,dimensiones,rfio,Q,cor_int,cor_ext,interno,tipo_flujo,par):
        self.cef=cef
        self.cec=cec
        self.csf=csf
        self.csc=csc
        self.dimensiones=dimensiones #pandas df columnas['Di', 'do', 'di', 'L','k']
        self.rfio=rfio
        self.Q=Q
        self.cor_int=cor_int
        self.cor_ext=cor_ext
        self.interno=interno
        self.tipo_flujo=tipo_flujo #'contraflujo' o 'coflujo'
        self.par=par   #Paralelos
        self.Re_Pr_int()
        self.Re_Pr_ext()
        self.U_calc()
        self.LMTD_calc()
        self.Num_int_calc()
        #self.dP_calc()
        #if self.operacion='ciclico':
         #   bababa
        #else:
         #   bububu
    def Re_Pr_int(self):
        import math
        self.Ai=math.pi*((self.dimensiones['di'][0]**2)/4)*self.dimensiones['Nt'][0]/self.dimensiones['Np'][0]
        self.Ae=(self.dimensiones['Ds'][0]*self.dimensiones['C'][0]*self.dimensiones['B'][0])/self.dimensiones['Pt'][0]
        if self.dimensiones['arreglo'][0]=='triangular' or self.dimensiones['arreglo'][0]=='triangular_rotado':
            self.Le_ext=4*((self.dimensiones['Pt'][0]**2)*((3**0.5)/4)-(math.pi*(self.dimensiones['do'][0]**2)/8))/(math.pi*self.dimensiones['do'][0]/2)
        else:
            self.Le_ext=4*(((self.dimensiones['Pt'][0]**2)-((math.pi*self.dimensiones['do'][0]**2)/4))/(math.pi*self.dimensiones['do'][0]))
            
        if self.interno=='caliente':
            self.Ge=self.cef.fm/self.Ae
            self.Gi=self.cec.fm/(self.Ai*self.par)
            self.Re_int=self.Gi*self.dimensiones['di'][0]/((self.cec.v+self.csc.v)/2)
            self.Re_ext=self.Ge*self.Le_ext/((self.cef.v+self.csf.v)/2)
            #(Re_i,Re_o,Pr_i,Pr_o,e,di,do,miu,miuw,miuc,dh,L,Ac,Pc,Pr_b,Pr_w,Tw,T_b,qw,k,h,r,c,Pe,beta,phi,Pi):

            self.Pr_ext=((self.cef.Cp+self.csf.Cp)/2)*1000*((self.cef.v+self.csf.v)/2)/((self.cef.k+self.csf.k)/2)
            self.Pr_int=((self.cec.Cp+self.csc.Cp)/2)*1000*((self.cec.v+self.csc.v)/2)/((self.cec.k+self.csc.k)/2)
            self.resultado_in=[self.cor_int((self.dimensiones['do'][0]-self.dimensiones['di'][0]),self.Re_int, self.Re_ext,self.Pr_int,self.Pr_ext, 5e-6,self.dimensiones['di'][0],self.dimensiones['do'][0],((self.cec.v + self.csc.v) / 2),((self.cec.v + self.csc.v) / 2),((self.cec.v + self.csc.v) / 2),self.dimensiones['di'][0]/4, self.dimensiones['L'][0],
                                                            (math.pi*self.dimensiones['Ds'][0]**2)/4,1,self.Pr_int,self.Pr_int,((self.cec.T+self.csc.T)/2),((self.cec.T+self.csc.T)/2),1,1,1,1,1,1,1,1,self.dimensiones['Pt'][0] )]
            self.Nu_int=self.resultado_in[0]
            if self.Nu_int=='No se recomienda usar' or self.Nu_int==None:
                self.Nu_int=None
                self.h_i=None
            else:
                self.h_i=self.Nu_int*((self.cec.k+self.csc.k)/2)/self.dimensiones['di'][0]



        else:
            self.Ge=self.cec.fm/self.Ae
            self.Gi=self.cef.fm/(self.Ai*self.par)
            self.Re_int=self.Gi*self.dimensiones['di'][0]/((self.cef.v+self.csf.v)/2)
            self.Re_ext=self.Ge*self.Le_ext/((self.cec.v+self.csc.v)/2)
            self.Pr_ext=((self.cec.Cp+self.csc.Cp)/2)*1000*((self.cec.v+self.csc.v)/2)/((self.cec.k+self.csc.k)/2)
            self.Re_int=self.Gi*self.dimensiones['di'][0]/((self.cef.v+self.csf.v)/2)
            self.Pr_int=((self.cef.Cp+self.csf.Cp)/2)*1000*((self.cef.v+self.csf.v)/2)/((self.cef.k+self.csf.k)/2)
            self.resultado_in=[self.cor_int((self.dimensiones['do'][0]-self.dimensiones['di'][0]),self.Re_int, self.Re_ext,self.Pr_int,self.Pr_ext, 5e-6,self.dimensiones['di'][0],self.dimensiones['do'][0],((self.cef.v + self.csf.v) / 2),((self.cef.v + self.csf.v) / 2),((self.cef.v + self.csf.v) / 2),self.dimensiones['di'][0]/4, self.dimensiones['L'][0],
                                                            (math.pi*self.dimensiones['Ds'][0]**2)/4,1,self.Pr_int,self.Pr_int,((self.cef.T+self.csf.T)/2),((self.cef.T+self.csf.T)/2),1,1,1,1,1,1,1,1,self.dimensiones['Pt'][0] )]
            self.Nu_int=self.resultado_in[0]
            if self.Nu_int=='No se recomienda usar' or self.Nu_int==None:
                self.Nu_int=None
                self.h_i=None
            else:
                self.h_i=self.Nu_int*((self.cef.k+self.csf.k)/2)/self.dimensiones['di'][0]
        return self.Ai,self.Gi,self.Re_int,self.Pr_int,self.resultado_in,self.Nu_int,self.h_i
    
    def Re_Pr_ext(self): #C es área libre de tubo
        import math
        self.Ae=(self.dimensiones['Ds'][0]*self.dimensiones['C'][0]*self.dimensiones['B'][0])/self.dimensiones['Pt'][0]
        if self.dimensiones['arreglo'][0]=='triangular' or self.dimensiones['arreglo'][0]=='triangular_rotado':
            self.Le_ext=4*((self.dimensiones['Pt'][0]**2)*((3**0.5)/4)-(math.pi*(self.dimensiones['do'][0]**2)/8))/(math.pi*self.dimensiones['do'][0]/2)
        else:
            self.Le_ext=4*(((self.dimensiones['Pt'][0]**2)-((math.pi*self.dimensiones['do'][0]**2)/4))/(math.pi*self.dimensiones['do'][0]))
            
        if self.interno=='caliente':
            self.Ge=self.cef.fm/self.Ae
            self.Re_ext=self.Ge*self.Le_ext/((self.cef.v+self.csf.v)/2)
            self.Pr_ext=((self.cef.Cp+self.csf.Cp)/2)*1000*((self.cef.v+self.csf.v)/2)/((self.cef.k+self.csf.k)/2)
            self.resultado_ext=[self.cor_ext((self.dimensiones['do'][0]-self.dimensiones['di'][0]),self.Re_int, self.Re_ext,self.Pr_int,self.Pr_ext, 5e-6,self.dimensiones['di'][0],self.dimensiones['do'][0],((self.cec.v + self.csc.v) / 2),((self.cec.v + self.csc.v) / 2),((self.cec.v + self.csc.v) / 2),self.dimensiones['di'][0]/4, self.dimensiones['L'][0],
                                                            (math.pi*self.dimensiones['Ds'][0]**2)/4,1,self.Pr_int,self.Pr_int,((self.cec.T+self.csc.T)/2),((self.cec.T+self.csc.T)/2),1,1,1,1,1,1,1,1,self.dimensiones['Pt'][0] )]
            self.Nu_ext=self.resultado_ext[0]
            if self.Nu_ext=='No se recomienda usar' or self.Nu_ext==None:
                self.Nu_ext=None
                self.h_o=None
            else:
                self.h_o=self.Nu_ext*((self.cef.k+self.csf.k)/2)/self.Le_ext
        else:
            self.Ge=self.cec.fm/self.Ae
            self.Re_ext=self.Ge*self.Le_ext/((self.cec.v+self.csc.v)/2)
            self.Pr_ext=((self.cec.Cp+self.csc.Cp)/2)*1000*((self.cec.v+self.csc.v)/2)/((self.cec.k+self.csc.k)/2)
            self.resultado_ext=[self.cor_ext((self.dimensiones['do'][0]-self.dimensiones['di'][0]),self.Re_int, self.Re_ext,self.Pr_int,self.Pr_ext, 5e-6,self.dimensiones['di'][0],self.dimensiones['do'][0],((self.cef.v + self.csf.v) / 2),((self.cef.v + self.csf.v) / 2),((self.cef.v + self.csf.v) / 2),self.dimensiones['di'][0]/4, self.dimensiones['L'][0],
                                                            (math.pi*self.dimensiones['Ds'][0]**2)/4,1,self.Pr_int,self.Pr_int,((self.cef.T+self.csf.T)/2),((self.cef.T+self.csf.T)/2),1,1,1,1,1,1,1,1,self.dimensiones['Pt'][0] )]
            self.Nu_ext=self.resultado_ext[0]
            if self.Nu_ext=='No se recomienda usar' or self.Nu_ext==None:
                self.Nu_ext=None
                self.h_o=None
            else:
                self.h_o=self.Nu_ext*((self.cec.k+self.csc.k)/2)/self.Le_ext
        return self.Ae,self.Le_ext,self.Ge,self.Re_ext,self.Pr_ext,self.resultado_ext,self.Nu_ext,self.h_o
    def U_calc(self):
        if (self.h_o==None and self.h_i ==None) or (self.h_o==None and self.h_i!=None) or (self.h_o!=None and self.h_i==None):
            P1=None
            P2=None
            P3=None
            self.R=None
            self.Ud=None
        else:
            import math
            P1=(1/self.h_o)
            P2=math.log(self.dimensiones['do'][0]/self.dimensiones['di'][0])*self.dimensiones['do'][0]/(2*self.dimensiones['k'][0])
            P3=((1/self.h_i))*(self.dimensiones['do'][0]/self.dimensiones['di'][0])
            P4=self.rfio
            self.R=P1+P2+P3
            self.Ud=1/(P1+P2+P3)
        return self.R,self.Ud
    
    def LMTD_calc(self):
        import math
        #Cálculo del LMTD
        dt1=self.cec.T-self.csf.T
        dt2=self.csc.T-self.cef.T
        if dt1==dt2:
            self.LMTD=dt1
        else:
            self.LMTD=(dt1-dt2)/(math.log(dt1/dt2))
        #Cálculo del factor de corrección para tubo carcasa
        if self.interno=='caliente': #interno se refiere a lo que pasa por adentro
            Ta=self.cef.T
            Tb=self.csf.T
            ta=self.cec.T
            tb=self.csc.T
        else:
            Ta=self.cec.T
            Tb=self.csc.T
            ta=self.cef.T
            tb=self.csf.T
        R=(Ta-Tb)/(tb-ta)
        P=(tb-ta)/(Ta-ta)
        if R!=1:
            alpha=((1-(R*P))/(1-P))**(1/self.dimensiones['Np'][0])
            S=(alpha-1)/(alpha-R)
            P1=2-(S*(R+1-((R**2)+1)**0.5))
            P2=2-(S*(R+1+((R**2)+1)**0.5))
            P3=(R-1)*math.log(P1/P2)
            P4=(((R**2)+1)**0.5)*math.log((1-S)/(1-(R*S)))
            self.F=P4/P3
        else:
            S=P/(self.dimensiones['Np'][0]-((self.dimensiones['Np'][0]-1)*P))
            self.F=S*(2**0.5)/((1-S)*math.log((2-(S*(2-2**0.5)))/((2-(S*(2+2**0.5))))))
        self.LMTDc=self.LMTD*self.F
        return self.LMTD,self.LMTDc,self.F
    def Num_int_calc(self):
        import math
        if self.Ud==None:
            self.Q_int=None
            self.Num_int=None
            self.Ao=None
            self.ojo=None
        else:
            self.Ao=self.dimensiones['do'][0]*math.pi*self.dimensiones['Nt'][0]
            self.Q_int=(self.Ud*self.LMTDc*self.Ao)/1000
            self.ojo=self.Q/self.Q_int
            if self.Q%self.Q_int==0:
                self.Num_int=int(self.Q/self.Q_int) #Q int es el de cada intercambiador de calor
            else:
                self.Num_int=int(self.Q/self.Q_int)+1
        return self.Ao,self.Q_int,self.Num_int,self.ojo

###################################################INTERCAMBIADOR TUBO - CARCASA##########################################
```

```python
class corrientemix: #LÍQUIDAA agua pura!!!!!
    def __init__(self, fm,fv,T, propiedades, propiedadesm,Tref):
        self.fm=fm
        self.fv=fv
        self.T=T
        self.propiedades=propiedades
        self.propiedadesm=propiedadesm
        self.Tref=Tref
        self.dens()
        self.ent()
        self.conduct()
        self.visc()
    #Cálculo de entalpía específica y total
    def ent(self):
        import pandas as pd
        import numpy as np
        
        self.Tf=(self.T - 273.15) * (9/5) + 32
        self.CP=0.0005389*self.Tf + 0.3996
        self.K=11.93
        if self.K !=11.80:
            self.F=0.05632*self.K + 0.3361
        self.Cpf=self.CP*self.F
        self.Cpp=self.Cpf*4.1868
        pc=pd.DataFrame({},columns=['cp','fracc'])
        #Se debe de calcular el Cp en knovel de cada gas individual, ahora suponiendo que ya fue calculado queda de la sieguiente manera:
        #J/mol-K
        self.Cpa=(self.propiedades['Cp_A'][0]+(self.propiedades['Cp_B'][0]*self.T)+(self.propiedades['Cp_C'][0]*self.T**2)+(self.propiedades['Cp_D'][0]*self.T**3))/self.propiedades['mm'][0]
        self.ha=(((self.propiedades['Cp_A'][0]*(self.T-self.Tref)+((self.propiedades['Cp_B'][0]/2)*(self.T**2-self.Tref**2))+((self.propiedades['Cp_C'][0]/3)*(self.T**3-self.Tref**3))+((self.propiedades['Cp_D'][0]/4)*(self.T**4-self.Tref**4)))/(self.propiedades['mm'][0])))
        self.hp=self.Cpp*(self.T-self.Tref)
        pc['h']=[self.hp,self.ha]
        pc['cp']=[self.Cpp,self.Cpa]
        pc['fracc']=[self.fpet,self.fagua]
        self.Cp=np.sum(pc['cp']*pc['fracc'])
        self.h=np.sum(pc['h']*pc['fracc'])
        self.h_tot=self.fm*self.h
        return self.h, self.Cp, self.h_tot
    #Cálculo de densidad
    def dens(self):
        import pandas as pd
        self.Tf=(self.T - 273.15) * (9/5) + 32
        self.SG= -0.0003644 *self.Tf + 0.9278  
        self.ra=(self.propiedades['r_A'][0]*(self.propiedades['r_B'][0]**(-(1-(self.T/self.propiedades['r_C'][0]))**self.propiedades['r_n'][0])))*(1/1000)*(1000000)
        self.rp=self.SG*self.ra
        self.ppet=self.propiedadesm['fraccpet'][0]*self.fv*self.rp
        self.pagua=self.propiedadesm['fraccagua'][0]*self.fv*self.ra
        tot=self.ppet+self.pagua
        self.fpet=self.ppet/tot
        self.fagua=self.pagua/tot        
        vv=pd.DataFrame({},columns=[])
        vv['dens']=[self.rp,self.ra]
        vv['fracc']=[self.fpet,self.fagua]
        dd=0
        for i in range(len(vv)):
            dd+=vv['fracc'][i]/vv['dens'][i]
        self.r=1/dd
        return self.ra,self.SG,self.r,self.fpet,self.fagua
    def conduct(self):
        import pandas as pd
        self.Tf=(self.T - 273.15) * (9/5) + 32
        self.ki=-2.177e-05*self.Tf + 0.07588
        self.kp=self.ki*5.6783
        self.ka=self.propiedades['k_A'][0]+(self.propiedades['k_B'][0]*self.T)+(self.propiedades['k_C'][0]*self.T**2)
        #Convertir a fracción másica
        self.ppet=self.propiedadesm['fraccpet'][0]*self.fv*self.rp
        self.pagua=self.propiedadesm['fraccagua'][0]*self.fv*self.ra
        tot=self.ppet+self.pagua
        self.fpet=self.ppet/tot
        self.fagua=self.pagua/tot
        mezcla=pd.DataFrame({},columns=[])
        mezcla['con']=[self.kp,self.ka]
        mezcla['fracc']=[self.fpet,self.fagua]
        dd=0
        for i in range(len(mezcla)):
            dd+=mezcla['fracc'][i]/(mezcla['con'][i]**2)
        self.k=dd**(-1/2)
        return self.k,self.fpet,self.fagua
    def visc(self):
        import pandas as pd
        self.Tc=self.T-273.15
        self.vc=1701.30807847*self.Tc**-0.83078668 
        self.vc=self.vc*1e-6
        self.vp=self.vc*self.r
        self.va=10**((self.propiedades['m_A'][0])+(self.propiedades['m_B'][0]/self.T)+(self.propiedades['m_C'][0]*self.T)+(self.propiedades['m_D'][0]*self.T**2))*1e-3
        vv=pd.DataFrame({},columns=[])
        vv['vis']=[self.vp,self.va]
        vv['fracc']=[0.85160631245623,1-0.85160631245623]
        import math
        dd=0
        for i in range(len(vv)):
            dd+=vv['fracc'][i]*math.log(vv['vis'][i])
        self.v=math.exp(dd)
        return self.v
```

```python
class corriente: #LÍQUIDAA agua pura!!!!!
    def __init__(self, fm, T, propiedades, Tref, tipo):
        self.fm=fm
        self.T=T
        self.propiedades=propiedades
        self.Tref=Tref
        self.tipo=tipo
        self.ent()
        self.dens()
        self.visc()
        self.conduct()
    #Cálculo de entalpía específica y total
    def ent(self):
        self.Tf=(self.T - 273.15) * (9/5) + 32
        if self.tipo=='NARANJOS':
            self.CP=0.00054*self.Tf + 0.4009
        elif self.tipo=='POZOLEO':
            self.CP=0.0005461*self.Tf + 0.4085
        elif self.tipo=='MARFO':
            self.CP=0.0005389*self.Tf + 0.3996
        elif self.tipo=='PAPALOAPAN':
            self.CP=0.0005407*self.Tf + 0.4018
        else:
            print('Inválido') 
        if self.tipo=='NARANJOS':
            self.K=11.83
        elif self.tipo=='POZOLEO':
            self.K=11.94
        elif self.tipo=='MARFO':
            self.K=11.93
        elif self.tipo=='PAPALOAPAN':
            self.K=11.90
        else:
            print('Inválido')
        if self.K !=11.80:
            self.F=0.05632*self.K + 0.3361
        self.Cpf=self.CP*self.F
        self.Cp=self.Cpf*4.1868
        self.h=self.Cp*(self.T-self.Tref)
        self.h_tot=self.fm*self.h
        return self.h, self.Cp, self.h_tot
    #Cálculo de densidad
    def dens(self):
        self.Tf=(self.T - 273.15) * (9/5) + 32
        if self.tipo=='NARANJOS':
            self.SG=-0.0003752 *self.Tf + 0.9244
        elif self.tipo=='POZOLEO':
            self.SG=-0.0003727 *self.Tf + 0.9018
        elif self.tipo=='MARFO':
            self.SG= -0.0003644 *self.Tf + 0.9278
        elif self.tipo=='PAPALOAPAN':
            self.SG=-0.0003749 *self.Tf + 0.9218
        else:
            print('Inválido')   
        self.r_a=(self.propiedades['r_A'][0]*(self.propiedades['r_B'][0]**(-(1-(self.T/self.propiedades['r_C'][0]))**self.propiedades['r_n'][0])))*(1/1000)*(1000000)
        self.r=self.SG*self.r_a
        return self.r_a,self.SG,self.r
    def conduct(self):
        self.Tf=(self.T - 273.15) * (9/5) + 32
        if self.tipo=='NARANJOS':
            self.ki=-2.177e-05*self.Tf + 0.07616
        elif self.tipo=='POZOLEO':
            self.ki=-2.284e-05*self.Tf + 0.07785
        elif self.tipo=='MARFO':
            self.ki=-2.177e-05*self.Tf + 0.07588
        elif self.tipo=='PAPALOAPAN':
            self.ki=-2.189e-05*self.Tf + 0.07635
        else:
            print('Inválido')
        self.k=self.ki*5.6783
        return self.k
    def visc(self):
        self.Tc=self.T-273.15
        if self.tipo=='NARANJOS':
            self.vc=5498.51750969*self.Tc**-1.35019544
        elif self.tipo=='POZOLEO':
            self.vc=438.40870052*self.Tc**-0.898179
        elif self.tipo=='MARFO':
            self.vc=1701.30807847*self.Tc**-0.83078668
        elif self.tipo=='PAPALOAPAN':
            self.vc=5061.57818139*self.Tc**-1.38572153
        else:
            print('Inválido') 
        self.vc=self.vc*1e-6
        self.v=self.vc*self.r
        return self.v,self.vc
```

```python
class corriente: #Crudo líquidooo!!!!!
    def __init__(self, fm, T, Tref):
        self.fm=fm
        self.T=T
        self.Tref=Tref
        self.ent()
        self.dens()
        self.visc()
        self.conduct()
    #Cálculo de entalpía específica y total
    def ent(self):
        self.Tc=self.T-273.15
        self.Cp=0.005548*self.Tc+3.386
        self.h=self.Cp*(self.T-self.Tref)
        self.h_tot=self.fm*self.h
        return self.h, self.Cp, self.h_tot
    #Cálculo de densidad
    def dens(self):
        self.Tc=self.T-273.15
        self.r=-1.109*self.Tc+1065
        return self.r
    def conduct(self):
        self.Tc=self.T-273.15
        self.k=-2.309e-06*self.Tc**2+0.0005245*self.Tc+0.3795
        return self.k
    def visc(self):
        self.Tc=self.T-273.15
        self.v=-2.456e-10*self.Tc**3+ 1.141e-07*self.Tc**2- 1.955e-05*self.Tc+0.001408
        return self.v
```
