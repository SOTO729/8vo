```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos extra√≠dos de la imagen:
data = {
    "Operating pressure (bar)": [50, 100, 150, 200, 250, 300],
    "Composite layers": [3, 4, 6, 8, 10, 12],
    "Total tank mass (kg)": [14.5, 17.6, 26.6, 35.9, 45.3, 55.0],
    "Total material cost (USD)": [201, 261, 398, 540, 685, 834]
}

# Crear DataFrame
df = pd.DataFrame(data)
print("Tabla de datos extra√≠dos de la imagen:")
print(df)

# Graficar los datos
plt.figure(figsize=(10, 6))

# Ejemplo de gr√°fico: 
# En este ejemplo graficamos la masa total del tanque y el costo total en funci√≥n de la presi√≥n operativa.
plt.plot(df["Operating pressure (bar)"], df["Total tank mass (kg)"],
         marker="o", linestyle="-", color="blue", label="Total tank mass (kg)")
plt.plot(df["Operating pressure (bar)"], df["Total material cost (USD)"],
         marker="s", linestyle="--", color="red", label="Total material cost (USD)")

plt.xlabel("Operating pressure (bar)")
plt.title("Type III Pressure Variation Example Summary")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
